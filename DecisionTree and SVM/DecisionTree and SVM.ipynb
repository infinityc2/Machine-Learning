```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import graphviz
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

iris_dataset = pd.read_csv('iris_dataset.csv')
iris_input = iris_dataset.drop('Class', axis=1)
iris_target = iris_dataset['Class']

clf = tree.DecisionTreeClassifier(random_state=100)
clf = clf.fit(iris_input, iris_target)
y_pred = clf.predict(iris_input)

print(confusion_matrix(iris_target, y_pred))
print(classification_report(iris_target, y_pred))
print(accuracy_score(iris_target,y_pred)*100)
```

```
Class = ['Iris-setosa','Iris-versicolor','Iris-virginica']
dot_data = tree.export_graphviz(clf, out_file=None,
           feature_names=list(iris_input),
           class_names=Class,
           filled=True, rounded=True,
           special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(iris_input,iris_target, test_size = 0.2)
model_train = clf.fit(x_train, y_train)
model_pred = model_train.predict(x_test)

print(confusion_matrix(y_test, model_pred))
print(classification_report(y_test, model_pred))
print(accuracy_score(y_test,model_pred)*100)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import graphviz
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

car_dataset = pd.read_csv("car_dataset.csv")

car_dataset['Buying_encoded'] = car_dataset['Buying'].map({'low':0,
'med':1, 'high':2, 'vhigh':3})

car_dataset['Maint_encoded'] = car_dataset['Maint'].map({'low':0,
'med':1, 'high':2, 'vhigh':3})

car_dataset['Doors_encoded'] = car_dataset['Doors'].map({'2': 2, '3': 3, '4': 4, '5more': 5})

car_dataset['Persons_encoded'] = car_dataset['Persons'].map({'2': 2, '4': 4, 'more': 5})

car_dataset['Lug_boot_encoded'] = car_dataset['Lug_boot'].map({'small':0,
'med':1, 'big':2,})

car_dataset['Safety_encoded'] = car_dataset['Safety'].map( {'low':0, 'med':1, 'high':2})

car_input = car_dataset[['Buying_encoded','Maint_encoded',
'Doors_encoded','Persons_encoded',
'Lug_boot_encoded','Safety_encoded']]

car_target = car_dataset['Class']

car_model = clf.fit(car_input, car_target)
car_pred = car_model.predict(car_input)

print(confusion_matrix(car_target, car_pred))
print(classification_report(car_target, car_pred))
print(accuracy_score(car_target,car_pred)*100)
```

```
from sklearn import svm
clsf = svm.SVC(kernel = 'rbf', gamma = 'scale', decision_function_shape='ovo')
svmAlg = clsf.fit(iris_input, iris_target)

predict = svmAlg.predict(iris_input)

print(confusion_matrix(iris_target, predict))
print(classification_report(iris_target, predict))
print(accuracy_score(iris_target, predict)*100)
```
