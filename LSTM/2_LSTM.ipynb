```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles

from keras.layers import Dense, LSTM, Reshape, CuDNNLSTM
from keras.models import Sequential
```

```
# Define number of sample point in circle
n_sample = 2000

# Define number of feature for developing model
n_feature = 250
```

```
# Generate Dataset
circle = make_circles(n_samples=n_sample, shuffle=False)[0][:int(n_sample/2)]

plt.figure(figsize=(8,8))
plt.plot(circle[:,0], circle[:,1], 'bo')
plt.show()
```

```
# Data Preprocessing

# Preprocess by slide data.
# n_feature of X using for predicting Y

x_train = circle[:,0]
y_train = circle[:,1]

y_train_temp = np.append(y_train[n_feature:], y_train[0:n_feature])

x_train_ap = np.append(x_train, x_train)
x_train_temp = []

for i in range(0, y_train.shape[0]):
  x_train_temp.append(x_train_ap[i:i+n_feature])

x_train_temp = np.asarray(x_train_temp)
x_train_temp.shape

print(y_train_temp.shape)
print(x_train_temp.shape)
```

```
# Develop LSTM model
model = Sequential()
model.add(Reshape((n_feature,1), input_shape=(n_feature, )))
model.add(CuDNNLSTM(units=132, return_sequences=True))
model.add(CuDNNLSTM(units=132))
model.add(Dense(1))
model.compile(loss='mse', optimizer='sgd')

model.fit(x_train_temp, y_train, epochs=100, shuffle=False, verbose=0)
model.summary()
```

```
model.evaluate(x_train_temp, y_train)
```

```
y_pred = model.predict(x_train_temp)

plt.figure(figsize=(8,8))
plt.plot(circle[:,0], y_pred, 'ro')
plt.plot(circle[:,0], circle[:,1], 'bo')
plt.show()
```
