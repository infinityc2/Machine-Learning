```python
import numpy as np
import pandas as pd
import gensim
```

```python
df = pd.read_csv('abstract.csv', error_bad_lines=False)
df.head()
```

```python
data = df['ABST'].values

# Example of data
data[0]
```

```python
abstract = []

for abst in data:
    temp = gensim.utils.simple_preprocess(abst)
    abstract.append(temp)
    
abstract = np.asarray(abstract)
```

```python
# Generate Word2Vec model 
dim10 = gensim.models.Word2Vec(
        # Parameter of model.
        # Please read it's documents to find script of this part.
    abstract, size=5, window=5, min_count=1, workers=4
    
)

# Training part.
dim10.train(abstract, total_examples=len(abstract), epochs=10)
```

```python
# Defind 5 interested word.
words = ['deep', 'learning', 'rnn', 'lstm', 'gru']
```

```python
dim10.wv.most_similar(positive=words)
```

```python
# Generate Vector of Word
dim10[abstract[0]]
```
