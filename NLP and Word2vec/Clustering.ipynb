```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets, cluster

%matplotlib inline
```

```python
iris = datasets.load_iris()
iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris.head()
```

```python
# Select 2 features
feature1 = 'sepal length (cm)'
feature2 = 'sepal width (cm)'

#.....
data = iris
```

```python
plt.figure(figsize=(16,9))

fig = sns.scatterplot(
    x=feature1,
    y=feature2,
    data=data
    )

plt.title("Original iris data with selected 2 features")
plt.show(fig)
```

```python
# K-Mean
kmean_model = cluster.KMeans(n_clusters=3)
kmean_model.fit(data)

# DBSCAN Clustering
DBSCAN_model = cluster.DBSCAN(eps=0.1, min_samples=2)
DBSCAN_model.fit(data)

# Spectral Clustering
SpectralClustering_model = cluster.SpectralClustering(n_clusters=3)
SpectralClustering_model.fit(data)
```

```python
# Add clustering to another columns
#data = data.join(pd.DataFrame(data=kmean_model.predict(data), columns=['Kmean']))
data = data.join(pd.DataFrame(data=DBSCAN_model.labels_, columns=['DBSCAN']))
data = data.join(pd.DataFrame(data=SpectralClustering_model.labels_, columns=['SpectralClustering']))

data.head()
```

```python
plt.figure(figsize=(16,9))

fig2 = sns.scatterplot(
    x=feature1,
    y=feature2,
    hue='Kmean',
    style='Kmean',
    data=data
    )

plt.title("K-Mean")
plt.show(fig2)
```

```python
plt.figure(figsize=(16,9))

fig3 = sns.scatterplot(
    x=feature1,
    y=feature2,
    hue='DBSCAN',
    style='DBSCAN',
    data=data
    )

plt.title("DBSCAN")
plt.show(fig3)
```

```python
plt.figure(figsize=(16,9))

fig4 = sns.scatterplot(
    x=feature1,
    y=feature2,
    hue='SpectralClustering',
    style='SpectralClustering',
    data=data
    )

plt.title("Spectral Clustering")
plt.show(fig4)
```

```python
# Clear clustering Column (Optional)
data = data.drop(columns=['Kmean', 'DBSCAN', 'SpectralClustering'])
```
